```python
import pandas as pd
```

```python
from google.colab import drive
drive.mount('/content/drive')

file_path = "/content/drive/My Drive/cleaned_data.csv"


df1 = pd.read_csv(file_path)

file_path2 = "/content/drive/My Drive/dataincreased.csv"

df2 = pd.read_csv(file_path2)

df1.head()
```

```python
df1["order_date"] = pd.to_datetime(df1["order_date"])
df1["gmv"] = pd.to_numeric(df1["gmv"], errors="coerce")  # Convert to numeric
df1["deliverybdays"] = pd.to_numeric(df1["deliverybdays"], errors="coerce")
df1["deliverycdays"] = pd.to_numeric(df1["deliverycdays"], errors="coerce")
df1["order_id"] = df1["order_id"].astype("int64")
df1["order_item_id"] = df1["order_item_id"].astype("int64")
df1["pincode"] = df1["pincode"].astype("category")
```

```python
df1['Year'] = df1['order_date'].dt.year
df1['Month'] = df1['order_date'].dt.month

gmv_summary = df1.groupby(['Year', 'Month'])['gmv'].sum().reset_index()
gmv_summary.rename(columns={'gmv': 'Total_GMV'}, inplace=True)

print(gmv_summary.head())
```

```python
df2.head()
```

```python
df_analysis = pd.merge(gmv_summary, df2, on=['Year', 'Month'], how='inner')

print(df_analysis.head())
```

```python
df_analysis.info()
```

```python
df_analysis['Date'] = pd.to_datetime(df_analysis[['Year', 'Month']].assign(day=1)) + pd.offsets.MonthEnd(0)

# Drop 'Year' and 'Month' columns
df_analysis = df_analysis.drop(columns=['Year', 'Month'])

# Print updated dataframe
print(df_analysis.head())
```

```python
df_analysis.fillna(0, inplace=True)
```

```python
df_analysis.head()
```

```python
SEED = 105
```

```python
mdsp_cols=['TV','Digital','Sponsorship','Content Marketing','Online marketing',' Affiliates','SEM','Radio','Other']
control_vars=[]
```

```python
data_size = len(df_analysis)

n_media_channels = len(mdsp_cols)
n_extra_features = len(control_vars)
media_data = df_analysis[mdsp_cols].to_numpy()
extra_features = df_analysis[control_vars].to_numpy()
target = df_analysis['Total_GMV'].to_numpy()
costs = df_analysis[mdsp_cols].sum().to_numpy()
```

```python
# Split and scale data.
test_data_period_size = 24
split_point = data_size - test_data_period_size
# Media data
media_data_train = media_data[:split_point, ...]
media_data_test = media_data[split_point:, ...]
# Target
target_train = target[:split_point]
```

```python
media_scaler = preprocessing.CustomScaler(divide_operation=jnp.mean)
extra_features_scaler = preprocessing.CustomScaler(divide_operation=jnp.mean)
target_scaler = preprocessing.CustomScaler(divide_operation=jnp.mean)
cost_scaler = preprocessing.CustomScaler(divide_operation=jnp.mean, multiply_by=0.15)

media_data_train = media_scaler.fit_transform(media_data_train)

target_train = target_scaler.fit_transform(target_train)
costs = cost_scaler.fit_transform(costs)
```

```python
mmm = lightweight_mmm.LightweightMMM(model_name="hill_adstock")
```

```python
number_warmup=1000
number_samples=1000
```

```python
mmm.fit(
    media=media_data_train,
    media_prior=costs,
    target=target_train,
    number_warmup=number_warmup,
    number_samples=number_samples,
    media_names = mdsp_cols,
    seed=SEED)
```

```python
mmm.print_summary()
```

```python
plot.plot_media_channel_posteriors(media_mix_model=mmm, channel_names=mdsp_cols)
```

```python
plot.plot_prior_and_posterior(media_mix_model=mmm)
```

```python
plot.plot_model_fit(mmm, target_scaler=target_scaler)
```

```python
new_predictions = mmm.predict(media=media_scaler.transform(media_data_test),
                              seed=SEED)
```

```python
plot.plot_out_of_sample_model_fit(out_of_sample_predictions=new_predictions,
                                 out_of_sample_target=target_scaler.transform(target[split_point:]))
```

```python
plot.plot_media_baseline_contribution_area_plot(media_mix_model=mmm,
                                                target_scaler=target_scaler,
                                                fig_size=(30,10),
                                                channel_names = mdsp_cols
                                                )
```

```python
prices = jnp.ones(mmm.n_media_channels)
```

```python
n_time_periods = 20
budget = jnp.sum(jnp.dot(prices, media_data.mean(axis=0)))* n_time_periods
```
